# Customização 

```python
import matplotlib.pyplot as plt 
year = [1950, 1951, 1952, 2100]
pop = [2.538, 2.57, 2.62, 10.85]

#Add more data
year = [1800,1850,1900] + year 
pop = [1.0, 1.262, 1.650] + pop 

plt.plot(year,pop)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks([0,2,4,6,8,10], 
           ['0','2B','4B','6B','8B','10B']) #Ele modifica de onde iniciará o eixo Y e o nome 
plt.text(1950,2.1, 'Teste')

plt.grid(True) #Linha de grade 

plt.show()
```

# List vs Dictionary 

List = SELECT, UPDATE AND REMOVE:[]
- Sequência de valores, indexados por um intervalo de números 
- Coleção de valores, ordem é importante e selectionar facilmente subconjuntos interios de dados, usar lista.  
Dictionary = SELECT, UPDATE AND REMOVE:[]
- Indexado por chaves exclusivas, que podem ser de qualquer tipo imutável 
- Tabela de pesquisa, 0nde procurar dados e especificar chaves

## loc

```python
#Pegando apenas os países 
# 1 Colchete ele retorna uma série 
print(brics["country"])
print(type(brics["country"])) 
```

```python
# 2 Colchete ele retorna uma DataFrame 
print(brics[["country"]])
print(type(brics[["country"]])) 
```

```python
# Trazendo duas colunas no Pandas 
print(brics[["country","capital"]])
print(type(brics[["country","capital"]])) 
```

```python
#Retornando linhas específicas 
brics[1:4]
```

```python
# Usando loc 
# Formato Séries 
print(brics.loc["RU"])
print(type(brics.loc["RU"]))
```

```python
# Usando loc 
# Formato DataFrame  
print(brics.loc[["RU"]])
print(type(brics.loc[["RU"]]))
```

```python
# Usando loc 
# Formato DataFrame  
print(brics.loc[["RU","IN","SA"]])
print(type(brics.loc[["RU","IN","SA"]]))
```

```python
# Usando loc - Capturar Linhas e Colunas 
# Formato DataFrame  
brics.loc[["RU","IN","SA"], ["country","capital"]]
```

```python
# Usando Loc 
# Os dois  pontos voltam todas as linhas, mas apenas duas colunas 
brics.loc[:,["country","capital"]]
```
